# Car data analysis

```python
import pandas as pd
```

```python
df=pd.read_csv('cardata.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df=df.drop(['Engine Fuel Type','Market Category','Vehicle Style','Popularity','Number of Doors','Vehicle Size'],axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df=df.rename(columns={"Engine HP":"HP","Engine Cylinders":"Cylinders","Transmission Type":"Transmission","Driven_Wheels":"DriveMode","highway MPG":"HMPG","city mpg":"CMPG"})
```

```python
df.head()
```

```python
dup=df[df.duplicated()]
dup
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
df=df.dropna()
```

```python
df.info()
```

# finding IQR

```python
numVal = df.select_dtypes(include='number')

q3=numVal.quantile(0.75)
q1=numVal.quantile(0.25)

print(q3-q1)
```

# plots SNS and MATPLOTLIB

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(x=df['MSRP'])
```

```python
df.Make.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title("Number_of_cars_by_make")
plt.ylabel("Numbers_of_cars")
plt.xlabel("Make")
```

```python
plt.figure(figsize=(10,5))
c=df.corr()
sns.headmap(c,cmap="BrBG",annot=True,fmt=".2f")
```
